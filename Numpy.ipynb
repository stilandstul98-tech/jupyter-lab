```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def introduce(self):
        print(f"Hello, I {self.name}!")
        
    def have_bithday(self):
        self.age += 1
        print(f"This is my {self.age}!")
```

```python
student1 = Student("Oleg", 20)
student2 = Student("Anna", 22)
```

```python
student1.age, student2.name
```

```python
student2.introduce()
```

```python
student1.have_bithday()
```

```python
#создаем большой список
import time

size = 1_000_000
python_list = list(range(size))
numpy_array = np.arange(size)


#медленно
start = time.time()
result = [x * 2 for x in python_list]
print(f"Python цикл: {time.time() - start:.4f}sek")

#быстро
start = time.time()
result = numpy_array * 2
print(f"NumPy: {time.time() - start:.4f} sek")
```

```python
numpy_array[-5:], python_list[-5:]
```

```python
%%time
result = [x * 2 for x in python_list]
```

```python
%%time
start = time.time()
result = numpy_array * 2
print(f"NumPy: {time.time() - start:.4f} sek")
```
